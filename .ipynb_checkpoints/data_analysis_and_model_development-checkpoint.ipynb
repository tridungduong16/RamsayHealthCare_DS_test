# Part 1: Data Exploration and Preparation

## Handling Missing Values

In this step, we will identify columns in our dataset that contain missing values. Missing data can negatively affect analysis and model performance, so it's important to identify these columns early on.

We will use the `.isna().any()` method to detect if any missing values are present in each column. The `df.columns[df.isna().any()]` will return a list of column names where missing values exist, which we will print to better understand the scope of missing data in our dataset.


```python
missing_value_columns = df.columns[df.isna().any()]
print(f"Missing values in columns: {missing_value_columns}")
```

### Calculating Missing Values and Their Percentage

After identifying which columns contain missing values, the next step is to quantify them. 

In this step:
- We calculate the total number of missing values in each column using `.isna().sum()`, which counts the number of `NaN` values for the selected columns.
- Then, we calculate the percentage of missing values relative to the total number of rows in the dataset to understand the extent of missing data.
- Finally, we store this information in a new DataFrame with two columns: 
  1. **Missing Values** – the total count of missing entries in each column.
  2. **Percentage (%)** – the percentage of missing values in relation to the total number of rows in the dataset.

This allows us to identify whether the missing data is significant and requires further handling.


```python
missing_values_df = df[missing_value_columns].isna().sum()
total_rows = len(df)
missing_values_percentage = (missing_values_df / total_rows) * 100
missing_values_df = pd.DataFrame({
    'Missing Values': missing_values_df,
    'Percentage (%)': missing_values_percentage.round(2)
})
```

```python
missing_values_df.reset_index().rename(columns={"index": "Feature name"})
```

### Filling Missing Values with Grouped Median (AR-DRG)

In this step, we handle missing data by filling the missing values with the median of each group within the **AR-DRG** column. This approach ensures that the imputation is more contextually relevant, as it calculates the median separately for each **AR-DRG** group, instead of filling missing values with the global median. This method helps maintain the natural variability between different diagnosis-related groups, ensuring that the imputation does not introduce bias across different categories of diagnoses.

```python
def fill_median_per_group(df, group_col, fill_cols):
    df[fill_cols] = df.groupby(group_col)[fill_cols].transform(lambda x: x.fillna(x.median()))
    return df

df[missing_value_columns] = df[missing_value_columns].apply(pd.to_numeric, errors='coerce')
df = fill_median_per_group(df, 'AR-DRG', missing_value_columns)
```

# Outlier

```python
charge_columns = ['AccommodationCharge', 'CCU_Charges', 'ICU_Charge', 'TheatreCharge',
                'ProsthesisCharge', 'OtherCharges', 'BundledCharges']
df[charge_columns]
```

```python
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.8)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
```

```python
threshold = df['PharmacyCharge'].quantile(0.99)  # 99th percentile threshold
df['PharmacyCharge'] = df['PharmacyCharge'].apply(lambda x: threshold if x > threshold else x)
print(df['PharmacyCharge'].describe())
plt.figure(figsize=(10,6))
plt.hist(df['PharmacyCharge'], bins=50, color='skyblue')
plt.title('Distribution of PharmacyCharge (After Capping Outliers)')
plt.xlabel('PharmacyCharge ($)')
plt.ylabel('Frequency')
plt.yscale('log')  # Log scale to make the visualization clearer
plt.show()
```

# Part 2: Data Analysis and Visualisation

## Feature engineering

```python
df['AdmissionDate'] = pd.to_datetime(df['AdmissionDate'], errors='coerce')
df['SeparationDate'] = pd.to_datetime(df['SeparationDate'], errors='coerce')
df['LengthOfStay'] = (df['SeparationDate'] - df['AdmissionDate']).dt.days
df['LengthOfStay'] = df['LengthOfStay'].apply(lambda x: x if x >= 0 else None)
df['WeekendStay'] = df['AdmissionDate'].dt.weekday.apply(lambda x: 1 if x >= 5 else 0)
df['ICU_CCUse'] = df[['ICU_Charge', 'CCU_Charges']].notnull().any(axis=1).astype(int)
```

```python
bins = [0, 18, 40, 65, 100]
labels = ['Child', 'Young Adult', 'Adult', 'Elderly']
df['AgeGroup'] = pd.cut(df['Age'], bins=bins, labels=labels)
print(df[['Age', 'AgeGroup']].head())
```

## Visualization

```python
charge_columns = ['AccommodationCharge', 'CCU_Charges', 'ICU_Charge', 'TheatreCharge',
                'ProsthesisCharge', 'OtherCharges', 'BundledCharges']
df[charge_columns] = df[charge_columns].apply(pd.to_numeric, errors='coerce')
# df[charge_columns] = df[charge_columns].apply(lambda x: x.fillna(x.median()))
df['TotalCharges'] = df[charge_columns].sum(axis=1)
drg_charges = df.groupby('AR-DRG').agg(
    total_charges=pd.NamedAgg(column='TotalCharges', aggfunc='sum'),
    average_charges=pd.NamedAgg(column='TotalCharges', aggfunc='mean')
).reset_index()
top_drg_charges = drg_charges.sort_values(by='total_charges', ascending=False).head(10)
plt.figure(figsize=(10, 6))
sns.barplot(x='total_charges', y='AR-DRG', data=top_drg_charges, palette='viridis')
plt.xlabel('Total Charges ($)')
plt.ylabel('AR-DRG')
plt.title('Top 10 DRGs Accruing the Largest Total Charges')
plt.gca().invert_yaxis()  # Invert y-axis to show the largest at the top
plt.savefig('./diagram/top_drg_charges.pdf', format='pdf')
plt.show()
```

```python
conn = sqlite3.connect('test_database.db')
cursor = conn.cursor()
query = """
    SELECT
        strftime('%Y', AdmissionDate) AS year,
        strftime('%m', AdmissionDate) AS month,
        COUNT(*) AS total_admissions,
        AVG(COUNT(*)) OVER (PARTITION BY strftime('%Y', AdmissionDate), strftime('%m', AdmissionDate)) AS average_admissions
    FROM admissions_table
    WHERE AdmissionDate >= date('now', '-2 years')
    GROUP BY year, month
    ORDER BY year DESC, month DESC;
"""
cursor.execute(query)
rows = cursor.fetchall()
columns = ['year', 'month', 'total_admissions', 'average_admissions']
df_res = pd.DataFrame(rows, columns=columns)
conn.close()
df_res
```

```python
conn = sqlite3.connect('test_database.db')
cursor = conn.cursor()
query = """
WITH RankedCharges AS (
    SELECT 
        PrincipalDiagnosis,
        Sex,
        TotalCharges,
        ROW_NUMBER() OVER (PARTITION BY PrincipalDiagnosis, Sex ORDER BY TotalCharges) AS row_num,
        COUNT(*) OVER (PARTITION BY PrincipalDiagnosis, Sex) AS total_rows
    FROM 
        admissions_table
    WHERE 
        TotalCharges IS NOT NULL
)
SELECT
    PrincipalDiagnosis,
    Sex,
    MIN(TotalCharges) AS min_total_charges,
    MAX(TotalCharges) AS max_total_charges,
    AVG(CASE WHEN row_num = (total_rows + 1) / 4 THEN TotalCharges END) AS p25_total_charges,
    AVG(CASE WHEN row_num = (total_rows + 1) / 2 THEN TotalCharges END) AS median_total_charges,
    AVG(CASE WHEN row_num = 3 * (total_rows + 1) / 4 THEN TotalCharges END) AS p75_total_charges,
    COUNT(*) AS total_records
FROM 
    RankedCharges
GROUP BY 
    PrincipalDiagnosis, Sex
ORDER BY 
    PrincipalDiagnosis, Sex;
"""

cursor.execute(query)
rows = cursor.fetchall()
columns = ['PrincipalDiagnosis', 'Sex', 'min_total_charges', 'max_total_charges', 'p25_total_charges', 'median_total_charges', 'p75_total_charges', 'total_records']
df_res = pd.DataFrame(rows, columns=columns)
conn.close()
df_res
```

# Part 5: Model Development

## Total charge prediction

```python
target = 'TotalCharges'
numeric_features = ['Age', 'LengthOfStay']
categorical_features = ['AR-DRG', 'Principal_ProcedureCode', 'CareType', 'SourceOfReferral', 'UrgencyOfAdmission', 'AgeGroup', 'WeekendStay', 'Sex']
features = numeric_features +  categorical_features
X_train, X_test, y_train, y_test = train_test_split(df[features], df[target], test_size=0.3, random_state=42)
model_save_path="./models/"

numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])

preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])

models = {
    'RandomForest': RandomForestRegressor(random_state=42, n_estimators=50),
    'LinearRegression': LinearRegression(),
    'SVR': SVR(),
    'GradientBoosting': GradientBoostingRegressor(random_state=42, n_estimators=50),
}

results = []
for model_name, model_instance in models.items():
    model_pipeline = Pipeline(steps=[
        ('preprocessor', preprocessor),
        ('regressor', model_instance)
    ])
    model_pipeline.fit(X_train, y_train)
    model_filename = model_save_path + model_name + '_model.joblib'
    joblib.dump(model_pipeline, model_filename)
    print(f"Model saved: {model_filename}")
    y_pred = model_pipeline.predict(X_test)
    mae = mean_absolute_error(y_test, y_pred)
    rmse = mean_squared_error(y_test, y_pred, squared=False)
    r2 = r2_score(y_test, y_pred)
    mpe = mean_absolute_percentage_error(y_test, y_pred)
    results.append({
        'Model': model_name,
        'MAE': mae,
        'RMSE': rmse,
        'R² Score': r2,
        'MPE (%)': mpe
    })
results_df = pd.DataFrame(results)
```
